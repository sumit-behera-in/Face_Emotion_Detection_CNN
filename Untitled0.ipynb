```python
import pandas as pd
import os
import seaborn as sns
import glob
import numpy
```

```python
train_data='/content/images/train'
pd.DataFrame(os.listdir(train_data),columns=['Files_Name'])
```

```python
from keras.models import Sequential
from keras.layers import Conv2D , MaxPooling2D , Dense ,Dropout , Flatten ,BatchNormalization, Activation
from keras.optimizers import Adam,SGD
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_data_gen = ImageDataGenerator(
      rescale=1./255,
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')
validation_data_gen = ImageDataGenerator(
    rescale=1./255,
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')
```

```python
train_gen = train_data_gen.flow_from_directory(
    'images/train/',
    target_size=(48,48),
    batch_size=64,
    color_mode="grayscale",
    class_mode="categorical"
)
```

```python
valid_gen = validation_data_gen.flow_from_directory(
    'images/validation/',
    target_size=(48,48),
    batch_size=64,
    color_mode="grayscale",
    class_mode="categorical"
)
```

```python
from keras.losses import categorical_crossentropy
from keras import regularizers
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), padding='same', activation='relu', input_shape=(48, 48,1)))
model.add(Conv2D(64,(3,3), padding='same', activation='relu' ))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(128,(5,5), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D (pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(512,(3,3), padding='same', activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(512,(3,3), padding='same', activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(BatchNormalization())
model.add(MaxPooling2D (pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(512,(3,3), padding='same', activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(256,activation = 'relu'))
model.add(BatchNormalization())
model.add(Dropout(0.25))
model.add(Dense(512,activation = 'relu'))
model.add(BatchNormalization())
model.add(Dropout(0.25))
model.add(Dense(7, activation='softmax'))

model.compile(loss=categorical_crossentropy,optimizer=Adam(learning_rate=0.0001,weight_decay=1e-6),metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint("/content/drive/MyDrive/projects/face emotion/model1.h5", monitor='val_acc', verbose=1, save_best_only=True, mode='max')
callbacks_list = [checkpoint]

history = model.fit_generator(generator=train_gen,
                                steps_per_epoch=28821//64,
                                epochs=60,
                                validation_data = valid_gen,
                                validation_steps =7066//64,
                                callbacks=callbacks_list
                                )
```

```python
import matplotlib.pyplot as plt
h = history.history
h.keys()
```

```python
plt.plot(h['accuracy'])
plt.plot(h['val_accuracy'] , c = "red")
plt.title("acc vs v-acc")
plt.show()
```

```python
plt.plot(h['loss'])
plt.plot(h['val_loss'] , c = "red")
plt.title("loss vs v-loss")
plt.show()
```
